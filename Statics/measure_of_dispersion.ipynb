## Measure of dispersion

```python
import numpy as np
```

```python
ages_list=[23,43,23,56,74,32,68,98,45,32]
np.mean(ages_list)
```

## Variance

```python
var=np.var(ages_list)
```

## Standard Deviation

```python
import math
std=math.sqrt(np.var(ages_list))
print(std)
```

```python
np.std(ages_list)
```

```python
import seaborn as sns
sns.histplot(ages_list)
```

```python
sns.histplot(ages_list,kde=True)
```
